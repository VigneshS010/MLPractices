```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# apyori is not pre installed in google colab, so we need to istall it seperately
```

Importing dataset

```python

# In the datset header also have data, so we need to set as header= None

dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)

# print(dataset)
# Our model the data in the List format , so we need to convert it
# the model also requires that to be in string format

transactions = []
for i in range(0, len(dataset)):
  transactions.append([ str(dataset.values[i, j]) for j in range(0,20)])

# print(transactions)
```

Training the Apriori Model on the datset

```python
from apyori import apriori

# transactions it the list of out dataset
# min support = 3 * 7 / 7501, 3 products per day for week, the 7501 data for the week
# min confidence , 0.8 is mostly used but here 0.2
# min lift is good for more than 3
# min and max length - there only we want 2 products , one for buying and for free , buy and get one free
# if we want to set buy 10 A and get 1 B product free, min_length=1, max_length=11
rules = apriori(transactions = transactions, min_support = 0.003 , min_confidence =0.2, min_lift= 3, min_length=2,max_length = 2 )
```

Visualising the results

Displaying the first results coming directly from the output of the apriori function

```python
results = list(rules)
results
```

Putting the result well organised into a pandas DataFrame|

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
