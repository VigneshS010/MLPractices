Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Dataset

```python
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')

# We donts use matrix of feature or dependend variable
```

Implementing the Thompson Sampling

```python
import random
N = 500
d = 10
ads_selected = []

# Look at the formula
number_of_rewards_1 = [0] * d
number_of_rewards_0 = [0] * d

total_reward = 0

# Iterating through the all cycles
for n in range(0, N):
  ad = 0
  max_random = 0
  for i in range(0, d):
    random_beta = random.betavariate(number_of_rewards_1[i]+1, number_of_rewards_0[i]+1)

    # Storing the max randombeta value
    if random_beta > max_random:
      max_random = random_beta
      ad = i

  ads_selected.append(ad)
  reward = dataset.values[n, ad]
  if reward == 1:
    number_of_rewards_1[ad] += 1
  else:
    number_of_rewards_0[ad] += 1

  total_reward += reward

print("Ads selected:", ads_selected)
print("Number of rewards 0:", number_of_rewards_0)
print("Number of rewards 1:", number_of_rewards_1)
print("Total reward:", total_reward)


```

Visualising the result

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```

```python
# The thompson sampling finds the best ad within 500 cycles
# SO it considered as most efficient algo
```
