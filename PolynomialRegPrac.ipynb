Predicting the position of a person on his previous company based on the salary

Importing librairies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

Train and test split

```python
# In this MOdel we dont use train test split
# Because we use all data  for the training the model

# from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2, random_state =0)
```

Training Linear regression model

```python
# Linear reg we used to compare the accuracy with polynomial reg model
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

Training the Polnomial reg model

```python
from sklearn.preprocessing import PolynomialFeatures

# This creates the x1, x2^2, ..... x^n
# Changing the degree we can adjust the curve
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
# we pass x bcz it is the training data


# Now creating the b0, b1, b2 .... and adding to x1,x2^2.....
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
# fit requires 2 parameters x and y
```

Visualising the Linear regression result

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff')
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show()
```

Visualising the Polynomial regression result

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial reg)')
plt.xlabel ('Position level')
plt.ylabel('salary')
plt.show
```

More Smoother curve

```python
# This code gives more smoother curve

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial reg)')
plt.xlabel ('Position level')
plt.ylabel('salary')
plt.show()
```
